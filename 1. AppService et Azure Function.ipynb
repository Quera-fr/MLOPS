### Exercice 1 - Configuration du compte Azure


1. Créez un compte [Azure](https://azure.microsoft.com/fr-fr/free/).

2. Télécharger l'extension [Azure Tools](https://marketplace.visualstudio.com/items?itemName=ms-vscode.vscode-node-azure-pack) sur VSCode ainsi que l'application [Azure Core Tools](https://learn.microsoft.com/en-us/azure/azure-functions/functions-run-local?tabs=windows%2Cportal%2Cv2%2Cbash&pivots=programming-language-python)

3. Connectez VSCode à votre compte Azure.

4. Créez une fonction Azure Serverless à l'aide de l'outil Azure Functions.

-  Ouvrez la palette de commandes (Ctrl + Shift + P) et tapez Azure Functions: Create New Project.
-  Sélectionnez le dossier `api` pour votre projet.
-  Sélectionnez python comme langage de programmation.
-  Sélectionnez HTTP trigger comme modèle de déclencheur.
-  Entrez un nom pour la fonction : ISENFunction
-  Sélectionnez Anonymous comme niveau d'autorisation.

5. Modifier le code contenu dans le fichier `__init__.py`. Lorsq'une requette GET est déclenchée à partir de l'url d'une image, la fonction renvoie la prédiction de la classe de cette image, soit `cat` ou `dog`.


```python
import pickle


data = {'age': 58,
    'job': 1,
    'marital': 1,
    'education': 2,
    'default': 0,
    'balance': 2143,
    'housing': 1,
    'loan': 0,
    'campaign': 1,
    'pdays': -1,
    'previous': 0,
    'poutcome': 3
}


with open('model.pkl', 'rb') as f: model = pickle.load(f)
predict_value = model.predict(data)
predict_value
```

6. Déployez la fonction sur Azure.
- Modifier le fichier requirements.txt pour ajouter les packages nécessaires : `pandas` et `scikit-learn==1.3.1`.
- Rendez-vous dans l'onglet Azure de VSCode, faîtes sur la rubrique Functions App et sélectionnez Create New Function App in Azure.
- Entrez un nom unique pour votre application : ISEN
- Sélectionnez Python 3.7 comme version de Python.
- Sélectionnez le groupe de ressources ISEN
- Sélectionnez la région France Centre
- Une fois le déploiement terminé, faîtes un clic droit sur la fonction YnovFunction et sélectionnez `Start`.
- Faîtes un click droit sur le dossier `api` et sélectionnez deploy to function app puis sélectionnez la fonction nomISENApp.

### Exercice 2 - Déploiement d'une application Streamlit sur Azure

1. Créez une application Azure App Service à l'aide de l'extension Azure Tools intitulée nomIsenWebApp.

2. Modifier le fichier "app.py" pour faire une requette sur la fonction Azure créée à l'exercice 1.

3. Déployez l'application sur Azure.

4. Rendez-vous dans les configuration de votre apllication sur le portail Azure, onglet Configuration >> Paramètres généraux, puis ajoutez la commande de démarage suivante : `streamlit run app.py --server.port $PORT`

5. Testez votre application en vous rendant sur l'url de votre application.
